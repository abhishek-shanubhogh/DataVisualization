director A - 1,2,3,4

1,2,3,4 avg revenue -> 100

4-> risky if 4th budget - avg revenue(1,2,3,4) >=0

3-> risky if 3th budget - avg revenue(1,2,3,4) >=0

2-> risky if 2th budget - avg revenue(1,2,3,4) >=0

1-> risky if 1th budget - avg revenue(1,2,3,4) >=0

List the director names which have produced atleast one risky movie

```python
def risky(x):
  x['risky'] = x['budget'] - x["revenue"].mean()>=0
  return x
```

```python
data_risky = data.groupby("director_name").apply(risky)
data_risky
```

```python
data_risky.loc[data_risky["risky"]]
```

```python
data_agg = data.groupby("director_name").aggregate({"year":["min","max"],"title":["count"]})
data_agg
```

```python
data_agg.columns
```

```python
data_agg["year"]
```

```python
data_agg.columns = ["_".join(col) for col in data_agg.columns]
```

```python
data_agg.columns
```

```python
data_agg = data.groupby("director_name").aggregate(
        year_min= ("year","min"),
        year_max= ("year","max"),
        title_count = ("title","count")
        )
data_agg
```

```python
data_agg["year_duration"] = data_agg["year_max"] - data_agg["year_min"]
```

```python
data_agg["avg_movie_per_year"]= data_agg["title_count"] /data_agg["year_duration"]
```

```python
data_agg = data_agg.sort_values("avg_movie_per_year", ascending=False)
```
