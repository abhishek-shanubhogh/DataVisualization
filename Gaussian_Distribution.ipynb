```python
def get_cdf(height, point):
    return (height<=point).sum()/len(height)
```

```python
x_values = np.arange(50, 80, 0.1)
# x_values
```

```python
cdf_values = [get_cdf(height, point) for point in x_values]
```

```python
plt.plot(x_values, cdf_values)
plt.xlabel("height")
plt.ylabel("probability")
plt.show()
```
